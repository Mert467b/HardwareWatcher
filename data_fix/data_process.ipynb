```python
import pandas as pd
import re

def clean_row_value(val):
    
    if pd.isna(val) or isinstance(val, (int, float)):
        return val
    
    
    cleaned = re.sub(r'[^\d\.-]', '', str(val))
    
    try:
        return float(cleaned)
    except ValueError:
        return val 



df = pd.read_csv('../data_logs.csv', encoding='latin-1')


if 'cpu_boost_mode' in df.columns:
    df['cpu_boost_mode'] = df['cpu_boost_mode'].replace('aggresive', 'aggressive')


numeric_cols = [
    'cpu_temp_C', 'cpu_power_W', 'cpu_util_pct', 'cpu_freq_MHz', 
    'ram_used_GB', 'gpu_temp_C', 'gpu_util_pct', 'gpu_clock_MHz', 'gpu_power_W'
]

for col in numeric_cols:
    if col in df.columns:
        df[col] = df[col].apply(clean_row_value)


df.to_csv('data_logs_cleaned.csv', index=False)

print("Cleaning complete! Check 'data_logs_cleaned.csv'")
print(df.head()) 
```
